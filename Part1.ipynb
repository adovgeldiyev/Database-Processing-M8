```python
import pandas as pd
import numpy as np
```

```python
employees_pd = pd.read_csv("Employee.txt", names = ["FName", "MName", "LName",
                                                   "ID", "dateBirth", "Address",
                                                   "City", "State","Gender", "Salary", "ManagerID", "Hours"], 
                           header=None, index_col=False)
employees_pd.head()
```

```python
print("All male employees.")
employees_pd.loc[employees_pd['Gender']=='M']
```

```python
print("Lowest salary for female groups")
female_emp=employees_pd.loc[employees_pd["Gender"]=="F"]
female_emp['Salary'].min()
```

```python
middle_init = employees_pd["Salary"].groupby(employees_pd["MName"])
for name, salary in middle_init:
    print(f"Middle Initial: {name}\nSalary: {salary}\n---------")
```
